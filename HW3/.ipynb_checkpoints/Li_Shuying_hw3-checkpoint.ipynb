### Part 1: Read in the 'county_age_dist.csv' data and answer the next four questions.

__Exercise 3.1 (1 points).__ Cast the 'fips' column as a string such that it is a length 5 string of digits that starts with 0 if it is currently of length 4.  Yolo county for example is '06113'.  This is the age distribution of each county in the US (with territories), you may notice that the total population is unusually large, please ignore this.

```python
# read Data Set
countyAge = pd.read_csv('county_age_dist.csv')
countyAge.head()
```

```python
# Check types of the dataframe
countyAge.dtypes
```

```python
# Some lists will be use in latter exercises
age = countyAge.columns.tolist()[1:]
age
```

```python
# Write a function to change the type and value of column fips
def fips(x):
    """Return a length 5 string of digits that starts with 0 if it is currently of length 4
    """
    x = str(x) #convert the int to str
    if len(x) == 4: #check if the length is 5
        x = '0' + x # Add '0' at the front
    return x
```

```python
# Apply the function row by row of the column fips
countyAge['fips'] = countyAge['fips'].apply(fips)
```

```python
# Check 1 - length is 5
all(countyAge['fips'].apply(len) == 5)
```

```python
# Check 2 - all are string
all(countyAge['fips'].apply(type) == str)
```

__Exercise 3.2 (2 points).__ Create a population column which is the sum of the age counts for all age bins.  For each age bin create a column which is proportion of the age bin in the county populations, so for example '0-17_rat' is the ratio of '0-17' to the whole population for the county.  Print out the FIPS code for the county with the largest proportion of 25-34 year olds? What is the FIPS code for the county with the smallest proportion of 85+ year olds?

**Create a population column which is the sum of the age counts for all age bins.**

```python
# Create a column that record the sum of population of each FIPS
countyAge['total'] = countyAge[age].sum(axis=1)
countyAge.head()
```

```python
# Create rat columns name
cols_rat = [(lambda x:x+'_rat')(x) for x in age]
cols_rat
```

```python
# Combine existed columns + rat columns. The new column list will use as new columns of the dataframe
new_cols = countyAge.columns.tolist() + cols_rat
new_cols
```

```python
# Use .reindex to make new columns of the datafame, set fill value as 0
countyAge = countyAge.reindex(columns = new_cols, fill_value=0)
countyAge.head()
```

```python
# Use a for loop to calculate the proportion
for col_index in range(1,10):
    col_name = countyAge.columns[col_index]
    col_name_rat = countyAge.columns[col_index+10]
    countyAge[col_name_rat] = countyAge[col_name] / countyAge['total']

# Check if we successfully calculate the proportion
countyAge.head()
```

**Print out the FIPS code for the county with the largest proportion of 25-34 year olds?**

```python
# We first generate a list that contains all FIPS codes
# Then use .idxmax() to find the index of the largest proportion of 25-34 year olds
# Use that index to find the FIPS codes
countyAge['fips'].tolist()[countyAge['25-34_rat'].idxmax()]
```

The FIPS of the largest proportion of 25-34 year olds is 51013.

**What is the FIPS code for the county with the smallest proportion of 85+ year olds?**

```python
# We first generate a list that contains all FIPS codes
# Then use .idxmin() to find the index of the mallest proportion of 85+ year olds
# Use that index to find the FIPS codes
countyAge['fips'].tolist()[countyAge['85+_rat'].idxmin()]
```

The FIPS of the smallest proportion of 85+ year olds is 02158.

__Exercise 3.3 (1 points).__ What is the overall proportion of each age bin within the entire US.  You need to find the total number of each age bin across all counties and take the ratio with the total population.

```python
# Sum of population of the enitre US
# countyAge[age] will return a slice of dataframe that contains each age bins' values
# Then use .sum() to find the sum of all values from that slice of dataframe
us_populaiton = countyAge[age].values.sum()
us_populaiton
```

```python
# The overall proportion of each age bin within the entire US
# Total / entire us population
for age_bin in age:
    print(f'The overall proportion of {age_bin} age is: {round(countyAge[age_bin].sum()/us_populaiton,4) * 100}%')
```

__Exercise 3.4 (2 points).__ Create a new column that contains the mode of the age bins (the age bin with the largest proportion in the county. For example, the mode of FIPS 01001 is '0-17').  Print out the proportions of the modes over all counties in the US, so that the proportion of '0-17' is the percentage of the counties that have '0-17' as the largest age bin for that county.

**Create a new column that contains the mode of the age bins (the age bin with the largest proportion in the county**

```python
# Use .idxmax(axis=1) to find the column name that has largest proportion
# Since will return as age_bin + '_rat', we use .str[:-4] to delete the '_rat' part
countyAge['mode'] = countyAge[cols_rat].idxmax(axis=1).str[:-4]

# Check if successfully insert 
countyAge.head()
```

**Print out the proportions of the modes over all counties in the US**

```python
# Use value_counts() to check the proportion of modes 
mode_freq = countyAge['mode'].value_counts()
mode_freq
```

```python
# print out the propotion of modes
# frequency of each modes / total of numbers of counties
for i in range(len(mode_freq)):
    print(f'The percentages of modes of {age[i]} ages: {round(mode_freq[i]/countyAge.shape[0],5)*100}%')
```

__Exercise 3.6 (2 points)__ Drop any rows with missingness.  For each state calculate the proportion of each age bin as you did for the entire US in the question above.  What is the state with the most 18-24 proportion?  Give the proportions.

```python
# Drop if mode has null values
county_fips_master = county_fips_master[county_fips_master['mode'].notna()]
```

```python
# Drop if county_name has null values
county_fips_master = county_fips_master[county_fips_master['county_name'].notna()]
```

```python
# Check numbers of row
county_fips_master.shape
```

**For each state calculate the proportion of each age bin as you did for the entire US in the question above.**

```python
# Use groupby to find each state's population of each age bins
stateDF = county_fips_master[['state_name'] + age]
stateDF = stateDF.groupby('state_name').sum()
stateDF['total'] = stateDF[age].values.sum(axis=1)
stateDF = stateDF[age].div(stateDF["total"].values, axis=0)
stateDF
```

**What is the state with the most 18-24 proportion? Give the proportions.**

```python
# Find the state name which has the most 18-24 proportion
stateDF['18-24'].idxmax()
```

```python
# Calculate the 18-24 proportion
round(stateDF['18-24'].max()*100,4)
```

North Dakota is the state with the most 18-24 proportion which is 14.0018%.
